# DynamicDAG - Dynamic Styling Examples

This notebook demonstrates how to customize the appearance of your DAG widgets using dynamic styles.

## Example 1: Default Styles

Using the widget with default purple and green colors.

````python
from ipydagflow import DynamicDAG

nodes = [
    {"id": "db", "type": "datasource", "data": {"label": "Database"}},
    {"id": "api", "type": "datasource", "data": {"label": "API"}},
    {"id": "transform", "type": "box", "data": {"label": "Transform"}},
    {"id": "warehouse", "type": "datasource", "data": {"label": "Warehouse"}},
]

edges = [
    {"id": "e1", "source": "db", "target": "transform", "animated": True},
    {"id": "e2", "source": "api", "target": "transform", "animated": True},
    {"id": "e3", "source": "transform", "target": "warehouse"},
]

dag1 = DynamicDAG(nodes=nodes, edges=edges)
dag1
````

## Example 2: Custom Red & Blue Theme

````python
custom_styles = {
    "datasource": {
        "background": "rgba(255, 99, 132, 0.8)",  # Red
        "color": "white",
        "borderColor": "#ff6384",
        "borderWidth": 3,
    },
    "box": {
        "background": "rgba(54, 162, 235, 0.8)",  # Blue
        "color": "white",
        "borderColor": "#36a2eb",
        "borderWidth": 2,
    },
}

dag2 = DynamicDAG(nodes=nodes, edges=edges, styles=custom_styles)
dag2
````

## Example 3: Teal & Yellow Theme

````python
teal_yellow_styles = {
    "datasource": {
        "background": "rgba(75, 192, 192, 0.9)",  # Teal
        "color": "black",
        "borderColor": "#4bc0c0",
        "borderWidth": 4,
    },
    "box": {
        "background": "rgba(255, 205, 86, 0.9)",  # Yellow
        "color": "black",
        "borderColor": "#ffcd56",
        "borderWidth": 2,
    },
}

dag3 = DynamicDAG(nodes=nodes, edges=edges, styles=teal_yellow_styles)
dag3
````

## Example 4: Dark Theme

````python
dark_styles = {
    "datasource": {
        "background": "rgba(45, 55, 72, 0.95)",  # Dark gray
        "color": "#63b3ed",  # Light blue text
        "borderColor": "#4299e1",
        "borderWidth": 2,
    },
    "box": {
        "background": "rgba(26, 32, 44, 0.95)",  # Darker gray
        "color": "#68d391",  # Light green text
        "borderColor": "#48bb78",
        "borderWidth": 2,
    },
}

dag4 = DynamicDAG(nodes=nodes, edges=edges, styles=dark_styles)
dag4
````

## Example 5: Update Styles Dynamically

You can change styles after creating the widget!

````python
# Create a widget
dag5 = DynamicDAG(nodes=nodes, edges=edges)
dag5
````

````python
# Change to purple theme
dag5.styles = {
    "datasource": {
        "background": "rgba(159, 122, 234, 0.8)",
        "color": "white",
        "borderColor": "#9f7aea",
        "borderWidth": 3,
    },
    "box": {
        "background": "rgba(237, 100, 166, 0.8)",
        "color": "white",
        "borderColor": "#ed64a6",
        "borderWidth": 3,
    },
}
````

## Example 6: No Positions (Auto-Layout)

You can omit position coordinates and let the widget automatically layout your nodes!

````python
# Complex DAG without manual positioning
auto_nodes = [
    {"id": "source1", "type": "datasource", "data": {"label": "Source A"}},
    {"id": "source2", "type": "datasource", "data": {"label": "Source B"}},
    {"id": "source3", "type": "datasource", "data": {"label": "Source C"}},
    {"id": "process1", "type": "box", "data": {"label": "Process 1"}},
    {"id": "process2", "type": "box", "data": {"label": "Process 2"}},
    {"id": "combine", "type": "box", "data": {"label": "Combine"}},
    {"id": "output", "type": "datasource", "data": {"label": "Output"}},
]

auto_edges = [
    {"id": "e1", "source": "source1", "target": "process1"},
    {"id": "e2", "source": "source2", "target": "process1"},
    {"id": "e3", "source": "source3", "target": "process2"},
    {"id": "e4", "source": "process1", "target": "combine"},
    {"id": "e5", "source": "process2", "target": "combine"},
    {"id": "e6", "source": "combine", "target": "output"},
]

orange_styles = {
    "datasource": {
        "background": "rgba(251, 146, 60, 0.85)",  # Orange
        "color": "white",
        "borderColor": "#f97316",
        "borderWidth": 2,
    },
    "box": {
        "background": "rgba(34, 211, 238, 0.85)",  # Cyan
        "color": "white",
        "borderColor": "#06b6d4",
        "borderWidth": 2,
    },
}

dag6 = DynamicDAG(nodes=auto_nodes, edges=auto_edges, styles=orange_styles)
dag6
````

## Style Configuration Reference

Each node type can be customized with the following properties:

```python
{
    "datasource": {  # or "box" or any custom type
        "background": "rgba(102, 126, 234, 0.8)",  # Background color (supports rgba)
        "color": "white",                           # Text color
        "borderColor": "#5a67d8",                  # Border color (hex or rgba)
        "borderWidth": 2,                           # Border width in pixels
    }
}
```

**Tips:**
- Use `rgba()` for transparency (last value is opacity: 0-1)
- Hex colors work too: `#ff6384`
- Adjust `borderWidth` for thicker/thinner borders
- Selected nodes automatically get a blue highlight border
