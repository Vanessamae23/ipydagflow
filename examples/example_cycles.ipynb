# Graphs with Cycles

ipydagflow supports directed graphs with cycles. Here are some examples.

```python
from ipydagflow import Step, StepDAG
```

## Example 1: Simple Cycle

A simple cycle where step A leads to B, and B leads back to A.

```python
step_a = Step(id="a", label="Process A", step_type="box")
step_b = Step(id="b", label="Process B", step_type="box")

# Create a cycle: A -> B -> A
step_a.add_child(step_b)
step_b.add_child(step_a)

dag = StepDAG()
dag.add_steps(step_a, step_b)
dag.render()
```

## Example 2: Feedback Loop

A common pattern in machine learning and control systems.

```python
input_step = Step(id="input", label="Input Data", step_type="datasource")
process = Step(id="process", label="Process", step_type="box")
evaluate = Step(id="evaluate", label="Evaluate", step_type="box")
output = Step(id="output", label="Output", step_type="datasource")

# Linear flow with feedback
input_step.add_child(process)
process.add_child(evaluate)
evaluate.add_child(output)
evaluate.add_child(process)  # Feedback loop!

dag = StepDAG()
dag.add_steps(input_step, process, evaluate, output)
dag.render()
```

## Example 3: Iterative Training Loop

A machine learning training loop with model updates.

```python
data = Step(id="data", label="Training Data", step_type="datasource")
initialize = Step(id="init", label="Initialize Model", step_type="box")
train = Step(id="train", label="Train Epoch", step_type="box")
validate = Step(id="validate", label="Validate", step_type="box")
check_convergence = Step(id="check", label="Check Convergence", step_type="box")
save_model = Step(id="save", label="Save Model", step_type="datasource")

# Build the training loop
data.add_child(initialize)
initialize.add_child(train)
train.add_child(validate)
validate.add_child(check_convergence)
check_convergence.add_child(train)  # Loop back if not converged
check_convergence.add_child(save_model)  # Exit loop if converged

dag = StepDAG()
dag.add_steps(data, initialize, train, validate, check_convergence, save_model)
dag.render()
```

## Example 4: Complex Cycle with Branching

A more complex graph with multiple cycles and branches.

```python
start = Step(id="start", label="Start", step_type="datasource")
branch_a = Step(id="branch_a", label="Branch A", step_type="box")
branch_b = Step(id="branch_b", label="Branch B", step_type="box")
merge = Step(id="merge", label="Merge", step_type="box")
decision = Step(id="decision", label="Decision", step_type="box")
end = Step(id="end", label="End", step_type="datasource")

# Create a complex flow
start.add_children(branch_a, branch_b)
branch_a.add_child(merge)
branch_b.add_child(merge)
merge.add_child(decision)
decision.add_child(end)
decision.add_child(branch_a)  # Cycle back to branch A
decision.add_child(branch_b)  # Cycle back to branch B

dag = StepDAG()
dag.add_steps(start, branch_a, branch_b, merge, decision, end)
dag.render()
```

## Example 5: Self-Loop

A step that references itself.

```python
retry = Step(id="retry", label="Retry on Failure", step_type="box")
success = Step(id="success", label="Success", step_type="datasource")

# Self-loop for retry logic
retry.add_child(retry)  # Retry references itself
retry.add_child(success)  # Eventually succeeds

dag = StepDAG()
dag.add_steps(retry, success)
dag.render()
```

## Example 6: Message Queue Pattern

A circular message processing pattern.

```python
queue = Step(id="queue", label="Message Queue", step_type="datasource")
consumer = Step(id="consumer", label="Consumer", step_type="box")
processor = Step(id="processor", label="Processor", step_type="box")
producer = Step(id="producer", label="Producer", step_type="box")

# Circular flow
queue.add_child(consumer)
consumer.add_child(processor)
processor.add_child(producer)
producer.add_child(queue)  # Back to queue

dag = StepDAG()
dag.add_steps(queue, consumer, processor, producer)
dag.render()
```

## Example 7: State Machine

A state machine with multiple possible transitions.

```python
idle = Step(id="idle", label="Idle", step_type="box")
running = Step(id="running", label="Running", step_type="box")
paused = Step(id="paused", label="Paused", step_type="box")
error = Step(id="error", label="Error", step_type="box")

# State transitions
idle.add_child(running)
running.add_children(paused, error, idle)  # Can pause, error, or stop
paused.add_children(running, idle)  # Can resume or stop
error.add_child(idle)  # Reset on error

dag = StepDAG()
dag.add_steps(idle, running, paused, error)
dag.render()
```

## Custom Styling for Cycles

You can customize the appearance of graphs with cycles.

```python
custom_styles = {
    "datasource": {
        "background": "rgba(255, 165, 0, 0.8)",  # Orange
        "color": "white",
        "borderColor": "#ff8c00",
        "borderWidth": 3,
    },
    "box": {
        "background": "rgba(138, 43, 226, 0.8)",  # Purple
        "color": "white",
        "borderColor": "#8a2be2",
        "borderWidth": 2,
    },
}

a = Step(id="a", label="A", step_type="box")
b = Step(id="b", label="B", step_type="box")
c = Step(id="c", label="C", step_type="box")

a.add_child(b)
b.add_child(c)
c.add_child(a)  # Cycle

dag = StepDAG(styles=custom_styles)
dag.add_steps(a, b, c)
dag.render()
```
